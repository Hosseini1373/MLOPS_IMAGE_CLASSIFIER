```python
from PIL import Image
import io
import time
import datetime
from flask import jsonify
import json
import base64
```

```python
def insert_image(db,image_name):
    im = Image.open("./"+image_name)
    image_bytes = io.BytesIO()
    im.save(image_bytes, format='JPEG')
    t_orig = time.time()
    t = datetime.datetime.fromtimestamp(t_orig).strftime('%c')
    test = db['MLOPS'].insert_one({
        "picture":image_bytes.getvalue(),
        "timestamp":t,
        "torig":t_orig
    })
    return test
```
